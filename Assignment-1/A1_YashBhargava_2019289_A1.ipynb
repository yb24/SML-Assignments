```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def graph(formula, x_lower_limit, x_upper_limit, y_upper_limit, graph_label):
   x = np.arange(x_lower_limit, x_upper_limit, 0.1)
   y = eval(formula)
   plt.ylim(0, y_upper_limit)
   plt.plot(x, y, label=graph_label)
   plt.legend()
   plt.show()
   plt.close()
```

```python
graph('(1/np.sqrt(2*np.pi))*np.exp(-0.5*((x-2)**2))', -1, 10, 0.5, 'y=P(x|w1)')
```

```python
graph('(1/np.sqrt(2*np.pi))*np.exp(-0.5*((x-5)**2))', -1, 10, 0.5, 'y=P(x|w2)')
```

```python
graph('np.exp(-0.5*(6*x-21))', -1, 10, 100, 'y=P(x|w1)/P(x|w2)')
```

```python
graph('( 1/(1+(x-3)**2) ) / ( 1/(1+(x-3)**2) + 1/(1+(x-5)**2) )', -1, 10, 1, 'P(w1|x) for a1=3, a2=5, b=1')
```
